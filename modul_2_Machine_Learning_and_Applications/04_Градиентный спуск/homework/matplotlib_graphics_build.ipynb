```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# создание полотна
fig = plt.figure()

# устанавливаем размер изображения полотна
fig.set_size_inches([7,5])

# рисуем график по списку по X индексы элементов 0, 1, 2, 3
plt.plot([1, 4, 9, 16])
```

```python
# рисуем 2 графика
fig = plt.figure()
fig.set_size_inches([9, 7])
# рисуем точки для графика 1 линии и второго
plt.plot([1, 4, 9, 16])
plt.plot([1, 8, 27, 64])
plt.scatter([1, 2, 3], [4, 9, 16])
plt.scatter([1, 2, 3],[8, 27, 64])
# рисуем график


plt.xlabel('время',fontsize=10, color='#FF6347')
plt.ylabel('млн.',fontsize=12, color='#FF6347')

# настройка подписей осей
plt.tick_params(colors='#696969', labelsize=10,
               length=5, width=2)

# настроим легенду(в списке передаем все линии и точки)
plt.legend(['Просто ЗП', 'ЗП в Data Science', 'Точки(ЗП)', 'Точки(DS)'], title='Обозначения')
# чтобы не выводить объект
plt.show()
```

### Рисуем sin()

```python


fig = plt.figure()
fig.set_size_inches([9, 7])

# генерируем 150 значений между -10 и 10
x = np.linspace(-10, 10, 150)
y = np.sin(x)

plt.plot(x,y)
plt.legend(['График sin()'], title='Обозначения')
plt.show()
```

### Несколько графиков на одном рисунке

```python
fig = plt.figure()

# размер картинки
fig.set_size_inches(15, 10)

# генерируем x вычисляем sin(x)
x = np.linspace(-10, 10, 150)
y = np.sin(x)

# первый график просим расположить в таблице 2 x 2 слева вверху
ax1 = fig.add_subplot(2, 2, 1)
plt.plot(x, y)

# аналогично, только просим расположить в правом верхнем углу
ax2 = fig.add_subplot(2, 2, 2)
plt.plot(x, y, color='brown')

# третий график расположим снизу на весь размер картинки, указав 2 строки 1 столбец, 2 строку
ax3 = fig.add_subplot(2, 1, 2)
plt.plot(x, y, color='green')

### Сделаем поприличнее расстояние между графиками
fig.tight_layout()
plt.show()
```
