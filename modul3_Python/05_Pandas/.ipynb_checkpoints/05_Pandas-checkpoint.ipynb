## Логическое ИЛИ

делается через символ `|` 

Оно требует выполнения хотя бы одного условия из всех.

```python
df[(df['humidity'] < 50) | (df['temp'] > 30.)] # найдем дни с низкой влажностью ИЛИ с высокой температурой,строго больше 30 градусов (30.)
```

## Логическое НЕ

Логическое "НЕ" делается через символ `~`

```python
df[~(df['workingday'] == 1)]

# выражение (df['workingday'] == 1) должно быть в скобках - иначе фильтрация НЕ не сработает
```

```python
df[
    (~(df['workingday'] == 0)) & (df['temp'] >= 10) & (df['temp'] < 30)
] 
# температура в диапазоне по выходным дням
# при фильтрации с несколькими значениями НЕ так же нужно окружить скобками
```

```python
# можно было использовать вместо ~ !=
df[
    ((df['workingday'] != 0)) & (df['temp'] >= 10) & (df['temp'] < 30)
]
```

## > Series и Index

### Series

`df['season'] == 1`

Это объект типа Series. Его можно воспринимать как DataFrame с одной колонкой. У Series есть индекс (англ. index) – числа слева – и значения, в данном случае, булевые (что можно увидеть как dtype: bool).

```python
df['season'] == 1

# то не True/False, а массив из True/False!

# происходило следующее: pandas искал в Series значения True, запоминал их индекс, 
# а затем в оригинальном датафрейме df забирал значения по этому индексу.
```

## Группировка

1. Отбираем подмножество по определенному критерию.
2. Над подмножеством применяем функцию, которая вернет одно число (т.е. превратит **целое подмножество** в **одно** число).
3. Возвращаемся в п.1, отбираем другое подмножество, повторяем алгоритм.
4. Возвращаем все полученные агрегации.

```python
"""
попросили посчитатать среднюю температуру в разбивке по сезонам.
Для этого надо сгруппировать по сезонам, т.е.
сделать 4 подмножества данных - по одному на каждый сезон,
затем каждую группу "схлопнуть" до одного числа - среднего по подмножеству.
Получим 4 числа - по одному среднему на каждую группу.
"""

# можно писать и так, возвращает Series, если не применить as_index=False
# df.groupby('season', as_index=False)['temp'].agg('mean')

df.groupby('season', as_index=False).agg({'temp': 'mean'})
```

_The aggregations that work with .agg() include:_

Mean - df.agg('mean')

Median - df.agg('median')

Mode - df.agg('mode')

Sum - df.agg('sum')

Count - df.agg('count')

Max - df.agg('max')

Min - df.agg('min')

_Standard Deviation - df.agg('std')_

Variance - df.agg('var')

Skewness - df.agg('skew')

Kurtosis - df.agg('kurt')

```python
# группировка по 2 параметрам, во всем df
df.groupby(['season', 'workingday']).mean()
```

```python
# а так только для одной колонки temp
df.groupby(['season', 'workingday'],as_index=False)['temp'].mean()
```

В последних двух примерах в полученных объектах DataFrame и Series индекс представлен объектом MultiIndex.

Работать с ним нетривиально, но его всегда можно сбросить через функцию `.reset_index()` или передав в функцию .groupby() ключ `as_index=False`

## Время и даты
pandas имеет свой nbg для работы с временем `pd.datetime`

```python
# тип колонки datetime до преобразования object
df.datetime[:3]
```

```python
# преобразование даты в pd.to_datetime
# datetime
df_1 = df.copy()
df_1['datetime'] = pd.to_datetime(df_1['datetime'])
df_1.datetime[:3]
```

```python
# можно использовать части даты для фильтрации, например месяц
df_1.loc[df_1.datetime.dt.month == 5][:3]
```

```python
# в исходном df дата не преобразована и фильровать по ней не получится
try:
    df.loc[df.datetime.dt.month == 5]
except:
    print('Отфильтровать не удалось')
```

```python
# группировка по дате
# сгрупировать данные по месяцам и вывести среднюю температуру и min влажность

df_1.groupby(df_1['datetime'].dt.month).agg({
    'temp': 'mean',
    'humidity': 'min'
})
```

Для даты можно узнать: ` день, месяц, день недели, номер недели, год`

документация :

https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#time-date-components

```python
# посчитали количество измерений по неделям
rents_by_week = df_1.groupby(df_1['datetime'].dt.weekofyear).agg({
    'temp': 'mean'
})

# выберем случайные 10 значений и зафиксируем результат
rents_by_week.sample(10, random_state=28)
```

### Гистограммы

`hist()` встроенная в pandas функция

`AxesSubplot` объект и к нему можно обращаться если положить в переменную, например ax, чтобы изменить график

```python
df_1['temp'].hist(figsize=(9,5), bins=100)
```

```python
ax = rents_by_week.hist(figsize=(9,5), bins=100)
ax[0, 0].set_title('Гистограмма понедельной $t^o$')
```

```python
ax = df_1.groupby(df_1['datetime'].dt.dayofyear,as_index=False).agg({
    'datetime': 'count'
}).hist(figsize=(9,5),
        bins=50
)

ax = ax[0, 0] # matplotlib возвращает двумерный массив, распакуем его
ax.set_title='Гистограмма количества поездок ежедневно'
ax.set_xlabel='Количество поездок'
ax.set_ylabel = 'Сколько раз было такое количество поездок'
```

### Точная настройка через matplotlib
можем самостоятельно отрисовать те же графики через `matplotlib`

```python
rents_by_week = df_1.groupby(df_1['datetime'].dt.weekofyear).agg({
    'datetime': 'count'
})
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(9,5))
ax.plot(rents_by_week)

# добавим еще точки на тот же график - этого не делали в pandas
ax.scatter(
    # по оси X - индекс. Наша неделя ушла в индекс после группировки
    rents_by_week.index,
     # по оси Y - значение колонки 0. Она называется temp, можно было по имени обратиться
    rents_by_week.iloc[:,0]
)
ax.set_title='Гистограмма количества поездок ежедневно'
ax.set_xlabel='Количество поездок'
ax.set_ylabel = 'Сколько раз было такое количество поездок'
ax.grid(True)
fig.show()
```
