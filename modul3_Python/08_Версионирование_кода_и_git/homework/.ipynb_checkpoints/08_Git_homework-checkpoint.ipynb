Мы попробуем смоделировать конфликт при merge: сделаем изменения в одной ветке, которые будут несовместимы с изменениями в другой ветке. Для начала мы эти изменения сделаем :)

`но прежде, нужно создать ветку dev (!) - клонируется только main`

- Теперь у нас есть две ветки, каждая из которых "предлагает" свое изменение над requirements.txt: в одной ветке второй строкой записано numpy, в другой - loguru.

- Сделайте git push обоих веток (можно перейти в каждую ветку и сделать git push, а можно сделать одним выстрелом через git push --all), потом сделайте ссылку с токеном и отправьте эту ссылку в поле ниже.

## 16

> Merge: учимся решать конфликты (3/4)

- Репозиторий: https://git.lab.karpov.courses/startml/git_hw_templates/merge_conflicts.git

- Откройте снова requirements.txt в состоянии конфликта. Будем его исправлять)

- В секции HEAD будет лежать "numpy" - так и должно быть, потому что на HEAD (сейчас это main) второй строкой прописан "numpy".

- В секции |||| предок будет пустота (=====) - потому что у dev и main общий предок - это коммит, где в requirements.txt лежит только одна строка "pandas" (второй строки нет). В этом вы можете убедиться, введя git log --graph --all .

- В секции >>>> dev будет лежать "loguru", потому что в версии от ветки dev второй строкой в requirements.txt выступает "loguru".

- Ознакомившись с тем, что пришло ото всех веток, мы принимаем решение включить все пакеты (и loguru, и numpy). Добейтесь этого: уберите все служебные записи, добавленные git, и приведите requirements.txt в корректный вид (он должен включать три пакета).

- Затем сообщите git, что проблема решена. Для этого добавьте requirements.txt в staging командой git add.

- Посмотрите на вывод git status. В нем должна быть строчка "All conflicts fixed but you are still merging" (или аналогичная на языке вашей ОС). Если это так, то можно делать коммит, который совершает merge, командой git commit.

- Сделайте коммит. Какое название дал git автоматически вашему коммиту? Не меняя этого имени, завершите коммит. Обратите внимание на комментарии: в них git написал вам, какие конфликты были.

`Merge branch 'dev'`
