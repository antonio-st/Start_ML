В этом `минипроекте` вам предстоит проанализировать данные из Google Play Store. Чуть ниже вы можете ознакомиться с описанием данных.

### Описание данных

<b> App  </b>— название приложения

<b> Category  </b>— категория, к которой относится приложение

<b> Rating </b> — рейтинг пользователей

<b> Reviews </b> — количество отзывов пользователей о приложении

<b> Size  </b>— размер приложения

<b> Installs  </b>— количество загрузок/установок приложения пользователями

<b> Type  </b>— платное или бесплатное приложение

<b> Price </b> — цена приложения

<b> Content Rating </b> — возрастная группа, на которую ориентировано приложение

<b> Genres  </b>— принадлежность приложения к нескольким жанрам

<b> Last Updated  </b>— дата последнего обновления приложения в Play Store

<b> Current Ver  </b>— текущая версия приложения в Play Store

<b> Android Ver </b> — минимальная требуемая версия Android

### 6
Научимся объединять датафреймы и познакомимся с функцией [concat](https://pandas.pydata.org/docs/reference/api/pandas.concat.html).

Допустим, мы хотим собрать датафрейм, который будет включать в себя: 
* первые три строки исходного датафрейма, строки 6-8 (включительно), 
* строки 16-19 (включительно) 
* и колонки: App, Size, Genres, Current Ver в соответствующем порядке.

По сути, это будет выглядеть `как объединение трех состоящих` из нескольких строк, датафреймов, являющимися срезами из исходных данных.

В качестве ответа загрузите полученный датафрейм в формате csv. При сохранении финального датафрейма, колонка с индексами должна остаться.<br>
*NB! Помните про индексацию с нуля и правила включения/невключения границ при слайсинге*

```python
# строки первые 3
ps_index_0_3 = playstore.loc[:2][['App','Size','Genres', 'Current Ver']]
# строки с 6 по 8
ps_index_6_8 = playstore.loc[5:7][['App','Size','Genres', 'Current Ver']]
# строки 16 по 19
ps_index_16_19 = playstore.loc[15:18][['App','Size','Genres', 'Current Ver']]
# concat - объединяем df
playstore_concate = pd.concat([ps_index_0_3, ps_index_6_8, ps_index_16_19])
# сохраняем в файл
playstore_concate.to_csv('playstore_concate.csv')
playstore_concate
```

```python
playstore.loc[:8]
```

```python
# альтернативный вариант
a = playstore[['App', 'Size', 'Genres', 'Current Ver']][:3] 
b = playstore[['App', 'Size', 'Genres', 'Current Ver']][5:8]
c = playstore[['App', 'Size', 'Genres', 'Current Ver']][15:19]
pd.concat([a, b, c]).to_csv('concate_playstore.csv')
pd.concat([a, b, c])
```

### 7
Для дальнейшего анализа нам необходимо убрать `дубликаты приложений` (колонка `App`) и сбросить индекс.

Сохраните полученный результат в виде датафрейма в переменную `unique_playstore`. Датафрейм уже находится в переменной playstore

[reset_index](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reset_index.html)<br>
[drop_duplicates](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html)

```python
unique_playstore = (playstore[(~(playstore['App'].duplicated()))]).reset_index(drop=True)
unique_playstore
```

### 8
В этом задании названия колонок необходимо привести к стандартному виду — все буквы должны быть нижнего регистра, а пробелы должны быть заменены на нижние подчеркивания. 

Методы, которые могут пригодиться: 

    rename
    str.lower
    str.replace
    columns

Данные уже сохранены в playstore. 

```python
playstore.columns = playstore.columns.str.lower().str.replace(' ', '_',regex = True)
```

### 9
Давайте посмотрим на долю платных и бесплатных приложений. В качестве ответа введите долю бесплатных приложений. Ответ округлите до сотых, в качестве десятичного разделителя используйте точку.

NB! Не забудьте убрать дубликаты перед подсчетом доли!

```python
# копия df
share_free_app = playstore.copy()
# ищем дубликаты приложений
share_free_app = share_free_app[(~(share_free_app.app.duplicated()))]
# используем value_counts(normalize=True) для вычисления доли
round(share_free_app.type.value_counts(normalize=True)['Free'], 2)
```

### 11
Давайте уберем лишние символы из колонки с ценой (price), чтобы далее с ней было удобнее работать, и переведем данные в тип float. <br>

*Данные уже сохранены в playstore.

```python
# меняем значения в столбце price: убираем знак $ и переводим тип во float
playstore['price'] = playstore.price.str.replace('$','', regex=True).astype(float)
```
